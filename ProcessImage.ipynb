```python
import os
os.environ["OPENCV_IO_MAX_IMAGE_PIXELS"] = str(pow(2,40))
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
PATH = 'Sample_Village/581304/581304_PANASPETA_GTIFF.tif'
img = cv.imread(PATH, cv.IMREAD_LOAD_GDAL)
```

```python
h,w,c = img.shape
print(h,w,c)
# plt.imshow(img)
# plt.title('my picture')
# plt.show()
```

```python
sub_img = []
stride = 4000
nh = int(h/stride)
nw = int(w/stride)
print(nh,nw)


start_col = 0; end_col = 0
for i in range(nh+1):
    end_col = end_col + stride
    start_row = 0; end_row = 0
    for j in range(nw+1):
        end_row = start_row + stride
        temp = img[start_row:end_row, start_col:end_col]
        image_path = f"images/581304/p{i}-q{j}.tif"
        if i<10:
            image_path = image_path.replace('p','0')
        else:
            image_path = image_path.replace('p','')
        if j<10:
            image_path = image_path.replace('q','0')
        else:
            image_path = image_path.replace('q','')
        
        print(image_path)
        cv.imwrite(image_path,temp)
        print("Width:",start_row,end_row, "Height:",start_col, end_col)
        sub_img.append(temp)
        # print(temp.shape)
        start_row = end_row
    start_col = end_col
    
# print(start_row,end_row)
# print(start_col, end_col)
temp = img[start_row:, start_col:]
sub_img.append(temp)
print(temp.shape)
```
